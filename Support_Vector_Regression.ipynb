```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('add.csv')
df.head()
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.head(10)
```

```python
df.shape
plt.figure(figsize=(9,9))
sns.heatmap(df.corr(),annot=True,cmap='Blues')
```

```python
X = df.drop(['Sales'],axis=1)
Y = df['Sales']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print(len(X_train),len(X_test))
```

```python
model=svm.SVR(kernel='linear')
model.fit(X_train,Y_train)
```

```python
preds=model.predict(X_train)
mse_score=metrics.mean_squared_error(Y_train,preds)
mse_score
```

```python
preds1=model.predict(X_test)
mse_score1=metrics.mean_squared_error(Y_test,preds1)
mse_score1
```

```python
def plot(X1=Y_test,Y1=preds1,title='Predicted vs Actual Values (Linear)'):
  plt.scatter(range(len(X1)),X1,c='Blue')
  plt.scatter(range(len(Y1)),Y1,marker='*',c='Red')
  plt.title(title)
  plt.show()
```

```python
plot()
```
